```python
import pandas as pd
import numpy as np

path = '/content/drive/MyDrive/Colab Notebooks/breast_cancer.csv'

df = pd.read_csv(path)

df
```

Linear Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

y = df['worst area']
X = df.drop('worst area', axis=1)

# 문자열 데이터가 있다면 제외하고 숫자 데이터만 사용
X = X.select_dtypes(include=[np.number])

# 데이터 분할 (75:25)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# 공통 오차 계산 함수들
def mse_np(actual, predicted):
    return np.mean((np.array(actual) - np.array(predicted)) ** 2)

def mse_pure(actual, predicted):
    sum_square_error = sum((a - p) ** 2 for a, p in zip(actual, predicted))
    return sum_square_error / len(actual)

model_lr = LinearRegression()
model_lr.fit(X_train, y_train)
pred_lr = model_lr.predict(X_test)

print("\n--- Linear Regression ---")
print('평균제곱근오차 (sklearn):', mean_squared_error(y_test, pred_lr))
print('평균제곱근오차 (numpy):', mse_np(y_test, pred_lr))
print('평균제곱근오차 (python):', mse_pure(y_test, pred_lr))
```

Decision Tree

```python
model_dt = DecisionTreeRegressor(random_state=0)
model_dt.fit(X_train, y_train)
pred_dt = model_dt.predict(X_test)

print("\n--- Decision Tree Regressor ---")
print('평균제곱근오차 (sklearn):', mean_squared_error(y_test, pred_dt))
print('평균제곱근오차 (numpy):', mse_np(y_test, pred_dt))
print('평균제곱근오차 (python):', mse_pure(y_test, pred_dt))
```

Random Forest

```python
model_rf = RandomForestRegressor(n_estimators=100, random_state=0)
model_rf.fit(X_train, y_train)
pred_rf = model_rf.predict(X_test)

print("\n--- Random Forest Regressor ---")
print('평균제곱근오차 (sklearn):', mean_squared_error(y_test, pred_rf))
print('평균제곱근오차 (numpy):', mse_np(y_test, pred_rf))
print('평균제곱근오차 (python):', mse_pure(y_test, pred_rf))
```
