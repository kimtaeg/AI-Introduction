#당뇨병 데이터


```python
import pandas as pd

#diabetes.csv파일 읽기
path = '/content/drive/MyDrive/Colab Notebooks/diabetes.csv'

#csv 파일을 읽어서 데이터로 만드는 코드
df = pd.read_csv(path)

df
```

1. 로지스틱 회귀(Logistic Regression)<br>
Logistic Regression은 데이터를 두 가지 범주로 분류하는 대표적인 머신러닝 모델이다.
입력 데이터의 특징들을 이용하여 특정 결과가 나올 확률을 계산하고, 그 확률을 기준으로 분류를 수행함.

```python
#데이터 불러오기
#df 데이터에서 Outcome 열만 꺼내서 y에 저장
y=df['Outcome']
#정답이 0이 몇 개 1이 몇 개인지 개수 확인
y.value_counts()
y

#outcome 열을 제외한 나머지 모든 열을 X로 만든 것
X=df.drop('Outcome',axis=1)
X.head()

from sklearn.model_selection import train_test_split

#데이터 쪼개기 75% 공부용, 25% 시험용
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=0)



from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

#모델 만들고 학습시키기
clf_lr = LogisticRegression(random_state=0)
clf_lr.fit(X_train, y_train)

#시험 데이터를 넣어 예측 및 결과 확인
pred_lr = clf_lr.predict(X_test)

print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(y_test, pred_lr))
print (confusion_matrix(y_test, pred_lr))
```

2. 의사결정나무(Decision Tree)<br>
Decision Tree는 조건을 반복적으로 나누면서 데이터를 분류하는 모델이다.
마치 사람이 의사결정을 할 때 “이 조건이 맞으면 이렇게, 아니면 저렇게”라고 판단하는 과정이랑 비슷함.


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# 모델 생성
clf_dt = DecisionTreeClassifier(random_state=0)

# 모델 학습. X_train는 문제 데이터, y_train는 정답
clf_dt.fit(X_train, y_train)

# 모델에게 새로운 데이터(X_test)를 넣어 예측
pred_dt = clf_dt.predict(X_test)

# 결과 출력
print("\n--- Decision Tree Classifier ---")
print("정확도(Accuracy):", accuracy_score(y_test, pred_dt))
print("혼동 행렬(Confusion Matrix):")
print(confusion_matrix(y_test, pred_dt))
```

3. 랜덤 포레스트(Random Forest)<br>
Random Forest는 여러 개의 Decision Tree를 만들어 그 결과를 종합하여 최종 결정을 내리는 앙상블 모델이다.
각각의 트리가 서로 다른 방식으로 학습한 후 다수결과 같은 방법으로 최종 예측을 수행함.



```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# 모델 생성. n_estimators=100는 의사결정나무를 100개 만든다는 뜻
# random_state=0는 실행할 때마다 결과 같게 만들기
clf_rf = RandomForestClassifier(n_estimators=100, random_state=0)

# 모델 학습
clf_rf.fit(X_train, y_train)

# 예측하기
pred_rf = clf_rf.predict(X_test)

# 결과 출력
print("\n--- Random Forest Classifier ---")
print("정확도(Accuracy):", accuracy_score(y_test, pred_rf))
print("혼동 행렬(Confusion Matrix):")
print(confusion_matrix(y_test, pred_rf))
```
